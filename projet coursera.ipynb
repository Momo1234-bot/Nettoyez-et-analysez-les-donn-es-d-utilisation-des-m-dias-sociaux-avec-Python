```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import random
```

```python
categories = ['spectacle', 'livre',]
n = 200
data = {
    'Date': pd.date_range('2021-01-01', periods=n),
    'Category': [random.choice(categories) for _ in range(n)],
    'Likes': np.random.randint(0, 10000, size=n)
}
```

```python
df = pd.DataFrame(data)
print(df.head())
print(df.info())
print(df.describe())
print(df['Category'].value_counts())
```

```python
df.dropna(inplace=True)
df.drop_duplicates(inplace=True)
df['Date'] = pd.to_datetime(df['Date'])
df['Likes'] = df['Likes'].astype(int)
```

```python
sns.histplot(df['Likes'], bins=30, kde=True)
plt.title('Distribution des Likes')
plt.xlabel('Likes')
plt.ylabel('Fr√©quence')
plt.show()
```

```python
sns.boxplot(x='Category', y='Likes', data=df)
plt.title('Boxplot des Likes par Cat√©gorie')
plt.xlabel('Cat√©gorie')
plt.ylabel('Likes')
plt.show()
```

```python
mean_likes = df['Likes'].mean()
print(f'Moyenne des Likes : {mean_likes}')
```

```python
mean_likes_by_category = df.groupby('Category')['Likes'].mean()
print(mean_likes_by_category)
```
